```python
#Project variables
project_id = 'realstate-381914'
table_id = 'AddDbd.Last3M'

```

```python
from google.cloud import bigquery
```

```python
client = bigquery.Client(project=project_id)
query = f'SELECT * FROM `{table_id}`  WHERE TIMESTAMP_TRUNC(_PARTITIONTIME, MONTH) = TIMESTAMP("2023-11-01") LIMIT 1000'

df = client.query(query).to_dataframe()

df.head(5)
```

```python
import pandas as pd
```

```python
df['block_timestamp'] = pd.to_datetime(df['block_timestamp']).dt.date
#df
```

```python
transaction_count = df.groupby('block_timestamp')['input_count'].sum().reset_index()
transaction_count
```
